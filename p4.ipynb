```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


y = np.array([1.49, 1.87, 1.86, 2.61, 2.43, 2.44, 3.00, 3.12])
x1 = np.array([100, 200, 300, 400, 500, 600, 700, 800])
x2 = np.array([26, 50, 26, 76, 51, 50, 75, 74])


plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
plt.scatter(x1, y, color="blue")
plt.xlabel("x1 (Fertilizer)")
plt.ylabel("y (Wheat Production)")
plt.title("Wheat Production (y) vs Fertilizer (x1)")


plt.subplot(1, 2, 2)
plt.scatter(x2, y, color="green")
plt.xlabel("x2 (Rainfall)")
plt.ylabel("y (Wheat Production)")
plt.title("Wheat Production (y) vs Rainfall (x2)")

plt.tight_layout()
plt.show()



correlation_x1_y = np.corrcoef(x1, y)[0, 1]
correlation_x2_y = np.corrcoef(x2, y)[0, 1]
print(f"Correlation between y and x1: {correlation_x1_y}")
print(f"Correlation between y and x2: {correlation_x2_y}")


X = np.column_stack((x1, x2))



model = LinearRegression().fit(X, y)
y_pred = model.predict(X)
residuals = y - y_pred

print(f"Regression coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
print(f"Residuals: {residuals}")


r_squared = r2_score(y, y_pred)
print(f"R-squared: {r_squared}")




```
